### Task 1
1. Write a function named is_prime that takes a number as a parameter and returns True if the number is prime and False otherwise. Print whether the number 29 is prime.

```python
def is_prime(n):
    # Check if the number is less than 2
    if n < 2:
        return False
    # Check for factors from 2 to the square root of n
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_status = is_prime(29)

# Test the function with the number 29
if prime_status:
    print(f"{29} is a prime number.")
else:
    print(f"{29} is not a prime number.")
```

### Task 2
1. Create a list of 10 numbers and sort them in descending order using the sort() function.
2. Use the len() function to print the length of the sorted list.
3. Use the zip() function to combine three lists (names, ages, and cities) into a list of tuples.
4. Use the range() function to create a list of even numbers from 2 to 20 and print the list.

```python
# Task 2

# Create a list of 10 numbers and sort them in descending order
numbers = [5, 3, 8, 1, 4, 10, 6, 2, 9, 7]
numbers.sort(reverse=True)

# Use the len() function to print the length of the sorted list
print(f"Length of the sorted list: {len(numbers)}")

# Create three lists and use the zip() function to combine them into a list of tuples
names = ["Ahmed", "Faris", "Khalid"]
ages = [25, 22, 22]
cities = ["Riyadh", "Al-Ghat", "Qassim"]

combined_list = list(zip(names, ages, cities))
print("Combined list of tuples (names, ages, cities):", combined_list)

# Use the range() function to create a list of even numbers from 2 to 20
even_numbers = list(range(2, 21, 2))
print("List of even numbers from 2 to 20:", even_numbers)
```

### Task 3
Build a phone book program that receives the phone number, and returns the name of the owner.
You can follow the table below:

| Name     | Number     |
| -------- | ---------- |
| Amal     | 1111111111 |
| Mohammed | 2222222222 |
| Khadijah | 3333333333 |
| Abdullah | 4444444444 |
| Rawan    | 5555555555 |
| Faisal   | 6666666666 |
| Layla    | 7777777777 |

- If the number exists, print the owner. Otherwise, print "Sorry, the number is not found".
- If the number is less or more than 10 numbers, print "This is invalid number".
- If the number contains letters or symbols, print "This is invalid number".


```python
# Phone book data as a dictionary
phone_book = {
    "Amal": "1111111111",
    "Mohammed": "2222222222",
    "Khadijah": "3333333333",
    "Abdullah": "4444444444",
    "Rawan": "5555555555",
    "Faisal": "6666666666",
    "Layla": "7777777777"
}

def find_owner(phone_number):
    # Check if the number is valid
    if len(phone_number) != 10 or not phone_number.isdigit():
        print("This is an invalid number.") # I think it should be This is an invalid input since it might be a string.
        return
    
    # Search for the owner of the number
    for name, number in phone_book.items():
        if number == phone_number:
            print(f"The owner of the number {phone_number} is {name}.")
            return
    
    print("Sorry, the number is not found.")

# Input from the user
user_input = input("Please enter the phone number: ")
find_owner(user_input)
```

### Task 4
- Given the following list : [5, 4, 17, 19, 30, 2, 7, 10, 45]
    1. Use a lambda function to filter out only the odd numbers.
    2. Print the new list of odd numbers.

```python
numbers = [5, 4, 17, 19, 30, 2, 7, 10, 45]

# Use a lambda function to filter out only the odd numbers
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))

print("List of odd numbers:", odd_numbers)
```
